```python
import matplotlib.pyplot as plt
from matplotlib.collections import EventCollection

import numpy as np
```

```python
def lagrange(arg, xs, ys):
    polynomial = 0

    for i in range(len(xs)):
        monomial = ys[i]
        for j in range(len(xs)):
            if i != j:
                monomial *= (arg - xs[j]) / (xs[i] - xs[j])
        polynomial += monomial

    return polynomial
```

```python
def f(x):
    return np.cos(x**2 + 2 * x)

a, b = 0, 1

step = 0.05

xs = np.array([0, 0.3, 0.5, 1], dtype=np.float64)
ys = f(xs)
```

```python
# table
print(*map(lambda p: f'{p:^ 4}', range(len(xs))), sep=' & ')
print(*map(lambda p: f'{p:^ 4}', xs), sep=' & ')
print(*map(lambda p: f'{p:^ 4}', ys), sep=' & ')
```

```python
x = np.arange(a, b + step, step) # lagrange x
y = lagrange(x, xs, ys) # lagrange y
```

```python
points = list(map(lambda p: f'({round(p[0], 3)}; {round(p[1], 3)})', zip(x,y))) # lagrange points

print(f'[len: {len(points)}]',*points)
```

```python
plt.plot(x, f(x), linewidth=.5, color='grey', label="cos(x^2 + 2x)")

plt.plot(xs, ys, ls='--', marker=".", mec="black", mfc="black", mew=5., color='grey', label="table [cos(x^2 + 2x)]")
plt.plot(x, y, color='black', label="lagrange")


plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```
